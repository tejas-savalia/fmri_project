```python
import pandas as pd
import numpy as np
```

```python
num_subjects = int(input("Number of Subjects: "))
```

```python
df = pd.DataFrame({"ID": np.zeros(num_subjects, dtype = object),
                   "FIRSTNAME": np.zeros(num_subjects, dtype = object),
                   "MIDDLENAME": np.zeros(num_subjects, dtype = object),
                   "LASTNAME":np.zeros(num_subjects, dtype = object),
                   "MOB":np.zeros(num_subjects, dtype = object),
                   "DOB":np.zeros(num_subjects, dtype = object),
                   "YOB":np.zeros(num_subjects, dtype = object),
                   "COB":np.zeros(num_subjects, dtype = object),
                   "SEX":np.zeros(num_subjects, dtype = object),
                   "SUBJECTHASMIDDLENAME":np.zeros(num_subjects, dtype = object)
                  })
```

```python
for subject in range(num_subjects):
    for column in df.columns:
        df[column][subject] = input(column + "value: ")
```

```python
df.loc[df.FIRSTNAME == 'Elijah', 'COB'] = 'Boston'
```

```python
df
```

```python
df.to_csv('guid_generator.csv', index = False)
```

## Race/Ethnicity/Gender breakdown

```python
guid_csv = pd.read_csv('guid_generator.csv')
```

```python
guid_csv.columns
```

```python
reg_csv = guid_csv[['SEX', 'Race', 'Ethnicity']].reset_index()
```

```python
reg_csv['identities'] = reg_csv['SEX']+', '+reg_csv['Race']+', '+reg_csv['Ethnicity']
reg_csv['identities'].str.split(', ', expland = True )
```

```python
reg_csv.groupby('identities').agg('count').reset_index().rename(columns = {'index':'counts'})[['identities', 'counts']]
```

```python
reg_csv.groupby(['SEX', 'Race', 'Ethnicity']).agg('count').reset_index().rename(columns = {'index': 'counts'})
```

```python
reg_csv.groupby(['Ethnicity']).agg('count').reset_index().rename(columns = {'index': 'counts'})[['Ethnicity', 'counts']]
```
